```python
import numpy as np
import matplotlib.pyplot as plt
from gravity3d import compute_gravity

# ---------------------------
# 1. Grid Parameters
# ---------------------------
nx, ny, nz = 30, 30, 15
Lx, Ly, Lz = 20e3, 20e3, 4e3  # in meters

x = np.linspace(-Lx/2, Lx/2, nx)
y = np.linspace(-Ly/2, Ly/2, ny)
z = np.linspace(0, Lz, nz)  # top to bottom
dx, dy, dz = x[1] - x[0], y[1] - y[0], z[1] - z[0]

x_edges = np.linspace(-Lx/2 - dx/2, Lx/2 + dx/2, nx + 1)
y_edges = np.linspace(-Ly/2 - dy/2, Ly/2 + dy/2, ny + 1)
z_edges = np.linspace(0 - dz/2, Lz + dz/2, nz + 1)

X, Y = np.meshgrid(x, y, indexing='ij')
Zc = z + dz/2  # Cell-centered depths

# ---------------------------
# 2. Define Gaussian Basin
# ---------------------------
depth_max = 3000  # meters
x0, y0 = 0.0, 0.0
sigma_x, sigma_y = 5e3, 5e3

Z_basin = depth_max * np.exp(-((X - x0)**2 / (2 * sigma_x**2) + (Y - y0)**2 / (2 * sigma_y**2)))

# ---------------------------
# 3. Create Density Model
# ---------------------------
rho_model = np.zeros((nx, ny, nz))
density_contrast = -500  # in kg/m³

for k in range(nz):
    mask = z[k] < Z_basin  # inside basin
    rho_model[:, :, k][mask] = density_contrast

# ---------------------------
# 4. Observation Grid
# ---------------------------
Xobs, Yobs = np.meshgrid(x, y, indexing='ij')
Zobs = np.zeros_like(Xobs)  # Observation at surface (z=0)

# ---------------------------
# 5. Compute Gravity Anomaly
# ---------------------------
gz = compute_gravity(Xobs, Yobs, Zobs, x_edges, y_edges, z_edges, rho_model)

# ---------------------------
# 6. Plot Gravity Anomaly
# ---------------------------
plt.figure(figsize=(8, 6))
contour = plt.contourf(X / 1e3, Y / 1e3, gz.T, 30, cmap='viridis')
plt.colorbar(contour, label='Gravity Anomaly (mGal)')
plt.title('Gravity Anomaly due to Gaussian Basin')
plt.xlabel('X (km)')
plt.ylabel('Y (km)')
plt.tight_layout()
plt.show()
```
